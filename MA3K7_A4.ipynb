```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
### Monte Carlo Simulation
# defining function that simulates game
def sim_walk(n, trials, start):
    '''Parameters: n is number we are testing for; trials is the number of simulations we run;
    start is the number we start from.
    Output: Returns the empirically 
    calculated probability of landing on n'''
    moves=[1,2]
    count = 0
    for i in range(trials):
        num = start
        while num<n:
            jump = random.choice(moves)
            num += jump
        if num == n:
            count += 1
    prob = count/trials
    return prob

#simulating game 100,000 times for n=25
print('Estimated probability for hitting 25 after 100,000 trials is:',sim_walk(25, 100000, 1))
```

```python
### Function calculating the exact solution
#defining function
def exact_walk(n):
    '''Parameters: n is number we are calculating probability of landing on
    Output: Returns the exact probability of landing on n'''
    return ((-1)**(n+1))*(2/3)*((1/2)**n) + (2/3)

# calculating exact probability of the game for n=25
print('The exact probability of landing on 25 when playing the game is:',exact_walk(25))
```

```python
## Plottting the probabilities of landing on each n>1 up to a maximum number
# parameters
max_n = 50

# calculating exact probabilities
n_vals = np.arange(1,max_n+1)
prob_vals = []
for k in n_vals:
    prob_vals.append(exact_walk(k))

# plotting results
plt.figure(figsize=(12,8))
plt.plot(n_vals, prob_vals, 'r')
plt.ylabel('Probability', size=17)
plt.xlabel('n', size=17)
plt.title('Probability of Landing on Each Integer n When Playing the Game', size=17)
plt.show()
```

```python
### Extension: Allowing Bigger Jumps
# defining function
def sim_walk2(n, trials, start, moves):
    '''Parameters: n is number we are testing for; trials is the number of simulations we run;
    start is the number we start from; moves is an array of numbers specifying the possible amounts
    we can move up the numberline in one move.
    Output: Returns the empirically 
    calculated probability of landing on n'''
    count = 0
    for i in range(trials):
        num = start
        while num<n:
            jump = random.choice(moves)
            num += jump
        if num == n:
            count += 1
    prob = count/trials
    return prob

# simulating for moves = [1,2,3] and n=25
print('The probability of landing on 25 when playing the game with moves=[1,2,3] is:',sim_walk2(25,100000, 1, [1,2,3]))

# simulating for n=25 and increasing number of moves
prob_data = []
x = np.arange(2,50)
for i in x:
    my_moves = np.arange(1,i)
    prob_data.append(sim_walk2(25,10000,1,my_moves))

# plotting results
plt.figure(figsize=(12,8))
plt.plot(x, prob_data, 'b')
plt.ylabel('Probability', size=17)
plt.xlabel('Maximum Possible Move', size=17)
plt.title('Probability of Landing on 25 Plotted Against the Maximum Possible Move Up the Number Line', size=16)
plt.show()
```
